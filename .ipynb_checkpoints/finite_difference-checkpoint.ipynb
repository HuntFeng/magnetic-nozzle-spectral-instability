```python
import numpy as np
from scipy.integrate import simpson
from nozzle import Spectral, Params, Nozzle, Boundary
import matplotlib.pyplot as plt
from tqdm import tqdm
```

# Dirichlet boundary

```python
N = 101
Mm = 1.0
constant_v = False
accelerating = True
boundary = Boundary.FIXED_FIXED

params = Params(Mm, constant_v, accelerating, boundary)
spectral = Spectral(N, "symmetric", "FD")
nozzle = Nozzle(params, spectral.x)
v0 = nozzle.v0
x = spectral.x
D1 = spectral.D1
D2 = spectral.D2

A2 = np.eye(*D1.shape)
A1 = 2j*(np.diag(v0)@D1 + np.diag(D1@v0))
A0 = np.diag(1-v0**2)@D2 \
        - np.diag((3*v0 + 1/v0)*(D1@v0))@D1 \
        - np.diag((1-1/v0**2)*(D1@v0)**2) \
        - np.diag((v0+1/v0)*(D2@v0))
V, nozzle.omega = nozzle.solve(A0[1:-1,1:-1],A1[1:-1,1:-1],A2[1:-1,1:-1])
nozzle.V = np.pad(V, ((1,1),(0,0))) # pad two ends by 0
nozzle.sort_solutions(real_range=[0,20], imag_range=[-2,2])

for i in range(nozzle.omega.size):
    v = nozzle.V[:,i]

    a = simpson(v.conj()*v,x)
    b = 2j*simpson(v.conj()*(v0*(D1@v) + (D1@v0)*v), x)
    c = simpson(
        v.conj()*((1-v0**2)*(D2@v) 
        - ((3*v0 + 1/v0)*(D1@v0))*(D1@v) 
        - ((1-1/v0**2)*(D1@v0)**2)*v 
        - ((v0+1/v0)*(D2@v0))*v)
        , x)
    roots = np.roots([a,b,c])
    if i == 0:
        omega = roots
    else:
        omega = np.row_stack([omega, roots])

fig, ax = plt.subplots(2,1, figsize=(7,5))
nozzle.plot_eigenvalues(ax=ax[0])
ax[0].plot(omega.real, omega.imag, "x", label="integral form")
ax[0].legend(fontsize="small")
ax[0].set_xlim([0,20])
nozzle.plot_eigenfunctions(num_funcs=3,ax=ax[1])
ax[1].legend(fontsize="small")
fig.suptitle(f"Finite Difference, $v_0={Mm}$")
plt.tight_layout()

# nozzle.save_data("FD",N=N)
```

## Change of variable 

Let $z=x^3$, then 
$$
\begin{align*}
\frac{d}{dz} &= 3x^{2/3}\frac{d}{dx} \\
\frac{d^2}{dz^2} &= 6x^{1/3}\frac{d}{dx} + 9x^{4/3}\frac{d^2}{dx^2}
\end{align*}
$$

```python
N = 101
Mm = 1.0
constant_v = False
accelerating = True
boundary = Boundary.FIXED_FIXED

params = Params(Mm, constant_v, accelerating, boundary)
spectral = Spectral(N, "symmetric", "FD")
nozzle = Nozzle(params, spectral.x)
x = spectral.x
# _D1 = spectral.D1
# _D2 = spectral.D2
D1 = spectral.D1
D2 = spectral.D2


# change of variable (numpy does not deal with fractional power)
v0 = nozzle.velocity_profile(x**3)
# D1 = np.diag(3*np.sign(x)*np.abs(x)**(2/3))@_D1
# D2 = np.diag(6*np.sign(x)*np.abs(x)**(1/3))@_D1 + np.diag(9*np.sign(x)*np.abs(x)**(4/3))@_D2

A2 = np.eye(*D1.shape)
A1 = 2j*(np.diag(v0)@D1 + np.diag(D1@v0))
A0 = np.diag(1-v0**2)@D2 \
        - np.diag((3*v0 + 1/v0)*(D1@v0))@D1 \
        - np.diag((1-1/v0**2)*(D1@v0)**2) \
        - np.diag((v0+1/v0)*(D2@v0))
V, nozzle.omega = nozzle.solve(A0[1:-1,1:-1],A1[1:-1,1:-1],A2[1:-1,1:-1])
nozzle.V = np.pad(V, ((1,1),(0,0))) # pad two ends by 0
nozzle.sort_solutions(real_range=[0,20], imag_range=[-5,5])

for i in range(nozzle.omega.size):
    v = nozzle.V[:,i]

    a = simpson(v.conj()*v,x)
    b = 2j*simpson(v.conj()*(v0*(D1@v) + (D1@v0)*v), x)
    c = simpson(
        v.conj()*((1-v0**2)*(D2@v) 
        - ((3*v0 + 1/v0)*(D1@v0))*(D1@v) 
        - ((1-1/v0**2)*(D1@v0)**2)*v 
        - ((v0+1/v0)*(D2@v0))*v)
        , x)
    roots = np.roots([a,b,c])
    if i == 0:
        omega = roots
    else:
        omega = np.row_stack([omega, roots])

fig, ax = plt.subplots(2,1, figsize=(7,5))
nozzle.plot_eigenvalues(ax=ax[0])
ax[0].plot(omega.real, omega.imag, "x", label="integral form")
ax[0].legend(fontsize="small")
ax[0].set_xlim([0,20])
for i in range(3):
    line = ax[1].plot(nozzle.x**3, nozzle.V[:,i].real/np.abs(nozzle.V[:,i].real).max(), label=f"$\omega=${nozzle.omega[i]:.3f}")
    ax[1].plot(nozzle.x**3, nozzle.V[:,i].imag/np.abs(nozzle.V[:,i].imag).max(), '--', color=line[-1].get_color())
    ax[1].set_xlabel("$z$")
    ax[1].set_ylabel("$\\tilde{v}$")
# ax[1].set_xlabel("x")
ax[1].legend(fontsize="small", loc="lower right")
fig.suptitle(f"Finite Difference, $v_0={Mm}$")
plt.tight_layout()

# nozzle.save_data("FD",N=N)
```

# Open right end 
Since the right end point is open, so the boundary point is constraint by interior points 
$$
0 = v_{n+1}' = \frac{3v_{n+1}-4v_n + v_{n-1}}{2h}
\Rightarrow
v_{n+1} = \frac{4}{3}v_n - \frac{1}{3}v_{n-1}
$$

This indicates that the derivatives at the last interior point should be modified as follows:
$$
v_{n}' = \frac{v_{n+1}-v_{n-1}}{2h} = \frac{1}{2h}\left(\frac{4}{3}v_{n} - \frac{4}{3}v_{n-1}\right) 
$$
$$
v_{n}'' = \frac{v_{n+1}-2v_{n}+v_{n-1}}{h^2} = \frac{1}{h^2}\left(-\frac{2}{3}v_{n} + \frac{2}{3}v_{n-1}\right) 
$$

```python
N = 301
Mm = 1.5
constant_v = False
accelerating = False
boundary = Boundary.FIXED_OPEN

params = Params(Mm, constant_v, accelerating, boundary)
spectral = Spectral(N, "symmetric", "FD")
nozzle = Nozzle(params, spectral.x)
v0 = nozzle.v0
x = spectral.x
dx = x[1] - x[0]
D1 = spectral.D1
D2 = spectral.D2
D1v = D1
D2v = D2
D1v[-2,:] = 0
D1v[-2,-3:-1] = np.array([-4/3, 4/3])/dx
D2v[-2,:] = 0
D2v[-2,-3:-1] = np.array([2/3, -2/3])/dx**2

A11 = np.zeros_like(D1)
A12 = np.eye(*D1.shape)
A21 = -np.diag(1-v0**2)@D2v \
        + np.diag((3*v0 + 1/v0)*(D1@v0))@D1v \
        + np.diag((1-1/v0**2)*(D1@v0)**2) \
        + np.diag((v0+1/v0)*(D2@v0))
A22 = -2j*(np.diag(v0)@D1v + np.diag(D1@v0))

A = np.block([[A11[1:-1,1:-1], A12[1:-1,1:-1]],[A21[1:-1,1:-1], A22[1:-1,1:-1]]])
V, nozzle.omega = nozzle.solve(A)
V = np.pad(V, ((1,1),(0,0))) # pad two ends by 0
V[-1,:] = (4*V[-2,:] - V[-3,:])/3 # modify the last row
nozzle.V = V
nozzle.sort_solutions(real_range=[-0.1,50], imag_range=[-1,1])

# for i in range(nozzle.omega.size):
#     v = nozzle.V[:,i]

#     a = simpson(v.conj()*v,x)
#     b = 2j*simpson(v.conj()*(v0*(D1@v) + (D1@v0)*v), x)
#     c = simpson(
#         v.conj()*((1-v0**2)*(D2@v) 
#         - ((3*v0 + 1/v0)*(D1@v0))*(D1@v) 
#         - ((1-1/v0**2)*(D1@v0)**2)*v 
#         - ((v0+1/v0)*(D2@v0))*v)
#         , x)
#     roots = np.roots([a,b,c])
#     if i == 0:
#         omega = roots
#     else:
#         omega = np.row_stack([omega, roots])

fig, ax = plt.subplots(2,1, figsize=(7,5))
nozzle.plot_eigenvalues(ax=ax[0])
# ax[0].plot(omega.real, omega.imag, "x", label="integral form")
# ax[0].legend(fontsize="small")
# ax[0].set_xlim([0, 50])
nozzle.plot_eigenfunctions(num_funcs=2,ax=ax[1])
ax[1].legend([f"$\omega_{n}$={nozzle.omega[n]:.3f}" for n in range(2)], fontsize="medium")
fig.suptitle(f"Finite Difference, $M_m={Mm}$")
plt.tight_layout()

# nozzle.save_data("FD",N=N)
```

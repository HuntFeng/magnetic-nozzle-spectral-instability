```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib_settting
```

```python
v0 = 0.5
N = 3

omega = np.arange(N+1)*np.pi*np.abs(1-v0**2)/2
z = np.linspace(-1,1,1000)
for n in range(1,N+1):
    v = np.exp(-1j*omega[n]/(v0+1)) * ( np.exp(1j*omega[n]*(z+1)/(v0+1)) - np.exp(1j*omega[n]*(z+1)/(v0-1)))
    lines = plt.plot(z,np.real(v),label=f"$\omega={omega[n]:.3f}$")
    plt.plot(z,np.imag(v),'--',color=lines[0]._color)
plt.xlabel("$z$")
plt.ylabel("$\\tilde{v}$")
plt.title(f"Exact Solutions, $v_0={v0}$")
plt.legend()
```
